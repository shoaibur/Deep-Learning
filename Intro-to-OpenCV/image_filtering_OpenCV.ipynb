# Creating a filters for image blurring and edge detection

### Import resources and display image

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,4), dpi=100)

# Read in the image
image = mpimg.imread('data/udacity_sdc.png')
ax[0].imshow(image)
ax[0].set_title('Original')

image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
ax[1].imshow(image, cmap='gray')
ax[1].set_title('Gray')

plt.show()
```

### Custom filters

```python
box_blur = 1/9 * np.array([[1,1,1],
                           [1,1,1],
                           [1,1,1]])
gauss_blur = 1/16 * np.array([[1,2,1],
                              [2,4,2],
                              [1,2,1]])
sobel_y = np.array([[-1,-2,-1],
                    [0,0,0,],
                    [1,2,1]])
sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,4), dpi=100)

imageGB = cv2.filter2D(image, -1, gauss_blur)
ax[0].imshow(imageGB, cmap='gray')
ax[0].set_title('Blur')

imageH = cv2.filter2D(imageGB, -1, sobel_y)
ax[1].imshow(imageH, cmap='gray')
ax[1].set_title('Horizontal edges')

imageV = cv2.filter2D(imageGB, -1, sobel_x)
ax[2].imshow(imageV, cmap='gray')
ax[2].set_title('Vertical edges')

plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,4), dpi=100)

imageGB = cv2.GaussianBlur(image, ksize=(3,3), sigmaX=0)
ax[0].imshow(imageGB, cmap='gray')
ax[0].set_title('Blur')

imageH = cv2.Sobel(imageGB, ddepth=cv2.CV_64F, ksize=3, dx=0, dy=1)
ax[1].imshow(imageH, cmap='gray')
ax[1].set_title('Horizontal edges')

imageV = cv2.Sobel(imageGB, ddepth=cv2.CV_64F, ksize=3, dx=1, dy=0)
ax[2].imshow(imageV, cmap='gray')
ax[2].set_title('Vertical edges')

plt.show()
```
